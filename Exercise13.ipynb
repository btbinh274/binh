# Creating a scatter plot

```python
import seaborn as sns
```

```python
# load a seaborn dataset
mpg_df = sns.load_dataset("mpg")

```

```python
mpg_df.head()
```

```python
# seaborn (‘version 0.9.0 is required’)
ax = sns.scatterplot(x="weight", y="mpg", data=mpg_df)
```

```python
ax = sns.scatterplot(x="weight", y="mpg", hue = "origin", data=mpg_df) # thêm trục thứ 3 theo màu sắc
```

```python
ax = sns.scatterplot(x="weight", y="mpg", hue = "origin", style = "cylinders", data=mpg_df)
# x, y: 2 trục dữ liệu
# hue: phân loại theo màu sắc: nên dùng biến định tính
# style: phân loại theo dấu marker, nên dùng biến định tính
```

```python
ax2 = sns.scatterplot(x="weight", y="horsepower", data=mpg_df)
```

```python
ax = sns.scatterplot(x="weight", y="mpg", hue = "displacement", style = "cylinders", data=mpg_df) #khi dung bien dinh luong cho hue se cho nhu vay
```
